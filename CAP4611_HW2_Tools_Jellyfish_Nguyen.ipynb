## P1: numpy array creation and printing

* Create a python list implementing a 5x5 array, with each element having , filled in with x^2 + y^2.
* Create from it the corresponding numpy array "arr"
* Print the shape of the array (should be 5 by 5)
* Print the type of the array (should be float64)

```python
# TODO: implement here
# source: https://www.geeksforgeeks.org/different-ways-to-create-numpy-arrays-in-python/#
arr = np.zeros((5,5))   # np.zeros() fills a matrix of n x m dimension

# is there already a way to do with with one code?
for i in range(5):
  for j in range(5):
    arr[i][j] = (i * i) + (j * j)

# source: https://www.geeksforgeeks.org/numpy-array-shape/
print("Array shape: ", arr.shape)
# source: https://www.geeksforgeeks.org/check-data-type-in-numpy/
print("Array type: ", arr.dtype)
```

## P2: array slicing and comparisons

* Extract the second column from arr
* Extract the second row from arr
* Check whether they are equal and print accordingly
* Create a new array arr2 which is a copy of the transpose of the original array.


```python
# TODO: implement here
# source: https://www.geeksforgeeks.org/how-to-access-a-numpy-array-by-column/
col2 = arr[:, 1]
row2 = arr[1, :]

# source: https://numpy.org/doc/stable/reference/generated/numpy.array_equal.html
if np.array_equal(col2, row2):
  print("The second column and second row are equal")
else:
  print("The second column and second row are not equal")

# creates arr2, a copy of arr's transpose
# source: https://numpy.org/doc/stable/reference/generated/numpy.transpose.html
# source: https://numpy.org/doc/stable/reference/generated/numpy.copy.html
arr2 = np.copy(np.transpose(arr))
```

## P4: solving a system of linear equations with numpy
* Consider a system of 17 linear equations.
* The parameters of the unknowns are the same as the array A created above.
* The left side of the equations is all 25.0 for each equation. Create a vector b for this.
* Solve this system of equations using numpy.linalg.solve. Print out the solutions.
* Solve the save system of equations, by taking the inverse of the A array and left multipying b with it. Print out the solutions.
* Check if the solutions are identical and print "Solutions are identical" if they are and "Solutions are not identical" if they are not. Discuss it in form of comments in the code.


```python
from os import X_OK
# TODO: implement here

# source: https://www.statology.org/numpy-fill-array/
vectorB = np.full((17,1), 25.0)

# source: https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html
x = np.linalg.solve(arrayA, vectorB)
print("System of Equations solved:")
print(x)

# source: https://www.geeksforgeeks.org/how-to-inverse-a-matrix-using-numpy/#
inverseA = np.linalg.inv(arrayA)
x2 = np.matmul(inverseA,vectorB)
print("Inverse of Systems of Equations solved:")
print(x2)

# source: https://numpy.org/doc/stable/reference/generated/numpy.allclose.html#numpy.allclose
# Using all close because the floats are probably too sensitive for normal np.equal()
if np.allclose(x,x2,1e-11):
  print("Solutions are identical")
else:
  print("Solutions are not identical")
```
